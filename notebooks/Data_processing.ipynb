```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import sys
import os
```

```python
players=['MATT','LIAM','TRAVIS','TALIESIN','SAM','LAURA','MARISHA','ASHLEY']
```

Load in the html files

# Dialo GPT Format

```python
attempt=[]
for i in os.listdir('dnd_transcript/'):
    if(i.split('.')[-1]=='html'):
        #print(i)
        with open(f"dnd_transcript/{i}",'r',encoding='latin-1') as f:
            html_text=f.read()
            soup=BeautifulSoup(html_text,'lxml')
            text=soup.get_text()
            for i in text.split('\n'):
                if(i.split(':')[0] in players):
                    attempt.append((i.split(':',maxsplit=1)[0],i.split(':',maxsplit=1)[1]))
contexted = []
n = 7
for i in range(n, len(attempt)):
    row = []
    prev = i - 1 - n # we additionally substract 1, so row will contain current responce and 7 previous responces  
    for j in range(i, prev, -1):
        row.append(attempt[j])
    contexted.append(row)       
final=[]
for i in contexted:
    final.append([i[0][0],i[0][1],i[1][1],i[2][1],i[3][1],i[4][1],i[5][1],i[6][1]])
data=pd.DataFrame(final,columns=[['player','response','context0','context1','context2','context3','context4','context5']])       
```

```python
data.to_csv('reponse.csv')
```

# DM to Player Model

```python
with open("Test1.html",'r',encoding='latin-1') as f:
    html_text=f.read()
```

```python
soup=BeautifulSoup(html_text,'lxml')
```

```python
text=soup.get_text()
```

```python
players=['MATT','LIAM','TRAVIS','TALIESIN','SAM','LAURA','MARISHA','ASHLEY']
```

```python
attempt=[]
for i in text.split('\n'):
    if(i.split(':')[0] in players):
        attempt.append((i.split(':',maxsplit=1)[0],i.split(':',maxsplit=1)[1]))
```

```python
dm_to_player=[]
for i in range(len(attempt)-1):
    if(attempt[i][0]=='MATT'):
        dm_to_player.append((attempt[i][0],attempt[i][1],attempt[i+1][0],attempt[i+1][1]))
        
```

```python
convo=pd.DataFrame(dm_to_player,columns=['DM','question','Player','Response'])
```

```python
final=convo[convo['Player']!='MATT'].reset_index(drop=True)
```
